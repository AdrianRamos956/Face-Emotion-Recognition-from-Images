```python
#here i am import everyhting important that we will need such as importing FER pandas and matplot
from fer import FER
import matplotlib.pyplot as plt
import pprint
from fer import Video
import os
import sys
import cv2
import pandas as pd
%matplotlib inline
```

```python
#This line of code gets the imaige and puts it inro its propperly names variable
test_image_one = plt.imread("Happy.jpeg")
#This is some initialization to call FER saying true
emo_detector = FER(mtcnn=True)
#This stores the captured emotions that are given when we send the imaige into the detect_emotions
captured_emotions = emo_detector.detect_emotions(test_image_one)
#This prints out all the emotions that we are given
print(captured_emotions)
#This prints out the imaige so we can actually see it
plt.imshow(test_image_one)
#This takes the imaige detection and gets the top emotion from it and tores it into dominant
dominant_emotion, emotion_score = emo_detector.top_emotion(test_image_one)
#it then prints out the dominant emotion for ease of use
print(dominant_emotion, emotion_score)
```

```python
#Puts the video into the right variable
location_videofile = "Video_One.mp4"
#then once again calls the FER and sets mtcnn to true
face_detector = FER(mtcnn=True)
#It then sends the video into the video function and puts that into the varable
input_video = Video(location_videofile)
```

```python
#This is where things really happen where this gets the video frame by frame and disects it so it can analyse it fame by frame, this takes a bit
processing_data = input_video.analyze(face_detector, display=False)
```

```python
#it then foes thorugh some professing to make sure we get the propper emotions from every frame
vid_df = input_video.to_pandas(processing_data)
vid_df = input_video.get_first_face(vid_df)
vid_df = input_video.get_emotions(vid_df)

#This lively funstion plots the vamues of the emotions in real time in a graph so we can see
pltfig = vid_df.plot(figsize=(20, 8), fontsize=16).get_figure()
```

```python
#Every frame that goes through updates these with the vallues of each emoition and adds it to the total
angry = sum(vid_df.angry)
disgust = sum(vid_df.disgust)
fear = sum(vid_df.fear)
happy = sum(vid_df.happy)
sad = sum(vid_df.sad)
surprise = sum(vid_df.surprise)
neutral = sum(vid_df.neutral)

#We then make a chart showing the total values that appear in the end so we know what the grates emotion that they felt over the span of the time
emotions = ['Angry', 'Disgust', 'Fear', 'Happy', 'Sad', 'Surprise', 'Neutral']
emotions_values = [angry, disgust, fear, happy, sad, surprise, neutral]

score_comparisons = pd.DataFrame(emotions, columns = ['Human Emotions'])
score_comparisons['Emotion Value from the Video'] = emotions_values
score_comparisons
```
